```python
from matplotlib import pyplot as plt
```

```python
# 2 bar graphs 
plt.bar([1,2,3,4,5],[50,40,70,80,20],label="BMW",width=0.5)

plt.bar([6,7,8,9,10],[80,40,50,60,20],label="Audi",color='g',width=0.5)
#legend() shows which colour of the plot represents which part
plt.legend()

plt.xlabel("no of years")
plt.xlabel("no of cars sold")
plt.title("cars sales data")
plt.show()

# horizontal graph

plt.barh([1,2,3,4,5],[50,40,70,80,20],label="BMW")
plt.legend()

plt.xlabel("no of years")
plt.ylabel("no of cars sold")
plt.title("cars sales data")
plt.show()
```

```python
# Histogram
# Number of repetition
from matplotlib import pyplot as plt

population=[20,27,28,22,20,27,65,65,43,13,55,33,22,49]

age_range=[0,10,20,30,40,50,100]

plt.hist(population,age_range,histtype='bar',rwidth=0.8)
plt.xlabel("Age Group")
plt.ylabel("Number of people")
plt.title("Histogram")
plt.show()
```

```python
from matplotlib import pyplot as plt

activities=['sleeping','eating','working','playing']

duration=[7,2,13,2]

colors=['r','b','g','y']

plt.pie(duration,labels=activities,colors=colors,shadow=True,autopct="%1.1f%%",explode=(0.2,0.2,0.5,0.1))

plt.title("My activities ")
plt.show()
```
