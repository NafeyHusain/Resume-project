```python
x=[1,2,3,4,5]

y=[11,21,31,41,51]
plt.title('Age vs Marks')
plt.ylabel('age')
plt.xlabel('marks')
plt.plot(x,y)

plt.show()
```

```python
x=[1,2,3,4,5]

y=[11,21,31,41,51]
plt.title('Age vs Marks')
plt.ylabel('age')
plt.xlabel('marks')
plt.scatter(x,y)
plt.show()
# colour and size
plt.scatter(x,y,c='g',s=150)

plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')

x1=[1,2,3,4,5]

y1=[11,21,31,41,51]

x2=[1,3,5,7,9]
y2=[3,6,9,12,15]


plt.title('Age vs Marks')
plt.ylabel('age')
plt.xlabel('marks')
plt.plot(x1,y1,c='b',label='x1 vs y1',linewidth=15)
plt.plot(x2,y2,c='g',label='x2 vs y2',linewidth=15)

plt.legend()
plt.grid(True,color='r')

plt.show()
```
